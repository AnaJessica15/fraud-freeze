```python
# https://app.gigasheet.com/datasets
df_address = pd.read_csv("/kaggle/input/ip-adressescsv-exported/ip_adresses.csv_exported.csv")
df_new = pd.concat([df, df_address], axis=1, join='inner')
```

```python
df_new.info()
```

```python
df_new['billing_latitude'] = np.radians(df_new['billing_latitude'])
df_new['billing_longitude'] = np.radians(df_new['billing_longitude'])
df_new['ip_address - Geo Latitude'] = np.radians(df_new['ip_address - Geo Latitude'])
df_new['ip_address - Geo Longitude'] = np.radians(df_new['ip_address - Geo Longitude'])
```

```python
df_new.head(15)
```

```python
df_new["dlon"] = df_new['ip_address - Geo Longitude'] - df_new['billing_longitude']
df_new["dlat"] = df_new['ip_address - Geo Latitude'] + df_new['billing_latitude']
y = df_new['ip_address - Geo Latitude'] - df_new['billing_latitude']
```

```python
df_new["dlat"] = 0.5 * (df_new["dlat"])
df_new["dlat"]=np.cos(df_new["dlat"])
x = (df_new["dlon"]) * (df_new["dlat"])
```

```python
R = 6371
df_new["dist"] = R * np.sqrt(x*x + y*y)
```

```python
drop_cols = ['billing_latitude','billing_longitude','ip_address - Geo Latitude','ip_address - Geo Longitude','dlon','dlat']

df_new.drop(drop_cols, axis = 1, inplace = True)
```

```python
# for ind in df_new.index:
#     print(df_new['is_fraud'][ind], df_new['ip_address - Geo Country Code'][ind], df_new['dist'][ind])
```

```python
df_new.loc[(df_new['billing_country'] == df_new['ip_address - Geo Country Code']) & (df_new['dist'] > 1000.0), 'address_verific'] = "low risk"  
```

```python
df_new.loc[(df_new['billing_country'] != df_new['ip_address - Geo Country Code']) & (df_new['dist'] > 11000.0), 'address_verific'] = "high risk" 
```

```python
df_new.loc[df_new['address_verific'].isnull(), 'address_verific'] = "legit"
```
